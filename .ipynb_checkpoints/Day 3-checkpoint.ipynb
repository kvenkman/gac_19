```python
filename = 'day3_input'
with open(filename, 'r') as f:
    lines = f.readlines()
```

```python
wire1 = lines[0]
wire2 = lines[1]

wire1 = wire1.split(',')
wire2 = wire2.split(',')
```

```python
def create_track(wire:list) -> list:
    current_coordinates = [0, 0]
    path = []
    
    path.append(current_coordinates)
    
    direction_dict = {'U': [0, 1], 'R': [1, 0], 'D': [0, -1], 'L': [-1, 0]}
    
    for trace in wire:
        direction = trace[0]
        
        
        step = direction_dict[direction]
        
            
        nsteps = int(trace[1:])
        
        
        for i in range(nsteps):
            current_coordinates[0] += step[0]
            current_coordinates[1] += step[1]
            
            path.append(current_coordinates) 
        #print(direction, step, nsteps, current_coordinates)
        print(path)
        break
    
    return path
```

```python
wire1_path = create_track(wire1)
```

```python
wire2_path = create_track(wire2)
```
