```python
filename = 'day5_input'
with open(filename, 'r') as f:
    line = f.readline()

line = line.split(',')
count = 0
myinput = str(5)

intcode_dict = {'01':4, '02':4, '03':2, '04':2, '05':3, '06':3, '07':4, '08':4, '99':0}
parameter_modes = {}
de = '00'
while(count < len(line)):
    #print("count, de: ", count, de)
    p_opcode = line[count]
    p_opcode = p_opcode.zfill(5)
    a, b, c, de = p_opcode[0], p_opcode[1], p_opcode[2], p_opcode[3:]
    
    
    if(de == '01'):
        #print(line[count], line[count + 1], line[count + 2], line[count + 3])
        parameter1 = line[int(line[count + 1])] if (c == '0') else line[count + 1]
        parameter2 = line[int(line[count + 2])] if (b == '0') else line[count + 2]
        parameter3 = int(line[count+3])
        #print(parameter1, parameter2, parameter3)
        
        line[parameter3] = str(int(parameter1) + int(parameter2))
        count += intcode_dict[de]
        continue
        
    if(de == '02'):
        parameter1 = line[int(line[count + 1])] if (c == '0') else line[count + 1]
        parameter2 = line[int(line[count + 2])] if (b == '0') else line[count + 2]
        parameter3 = int(line[count+3])
        
        line[parameter3] = str(int(parameter1) * int(parameter2))
        count += intcode_dict[de]
        continue
        
    if(de == '03'):
        parameter1 = int(line[count + 1])
        line[parameter1] = myinput
        count += intcode_dict[de]
        continue
        
    if(de == '04'):
        parameter1 = int(line[count + 1])
        print(line[parameter1])
        count += intcode_dict[de]
        continue

    if(de == '05'):
        parameter1 = line[int(line[count + 1])] if (c == '0') else line[count + 1]
        parameter2 = line[int(line[count + 2])] if (b == '0') else line[count + 2]
        
        #print(line[count + 1], line[count + 2], parameter1, parameter2, a, b, c, de, p_opcode, count)
        
        if(int(parameter1) != 0):
            count = int(parameter2)
        else:        
            count += intcode_dict[de]
        
        continue

    if(de == '06'):
        parameter1 = line[int(line[count + 1])] if (c == '0') else line[count + 1]
        parameter2 = line[int(line[count + 2])] if (b == '0') else line[count + 2]
        
        if(int(parameter1) == 0):
            count = int(parameter2)
        else:        
            count += intcode_dict[de]
        
        continue

    if(de == '07'):
        parameter1 = line[int(line[count + 1])] if (c == '0') else line[count + 1]
        parameter2 = line[int(line[count + 2])] if (b == '0') else line[count + 2]
        parameter3 = int(line[count+3])
        
        parameter1 = int(parameter1)
        parameter2 = int(parameter2)
        if(parameter1 < parameter2):
            line[parameter3] = '1'
        else:        
            line[parameter3] = '0'
            
        count += intcode_dict[de]
        continue

    if(de == '08'):
        parameter1 = line[int(line[count + 1])] if (c == '0') else line[count + 1]
        parameter2 = line[int(line[count + 2])] if (b == '0') else line[count + 2]
        parameter3 = int(line[count+3])

        parameter1 = int(parameter1)
        parameter2 = int(parameter2)
        if(parameter1 == parameter2):
            line[parameter3] = '1'
        else:        
            line[parameter3] = '0'
            
        count += intcode_dict[de]
        continue
        
    if(de == '99'):
        break
```
